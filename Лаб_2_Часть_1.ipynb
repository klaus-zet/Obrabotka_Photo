```python
import math
import numpy as np
from PIL import Image

img = Image.open('4Labs.png')
img_arr = np.array(img)
img_arr_2 = img_arr
width, height = img.size

def okno(x,y,z): 
    count = 0
    for i in range(3):
        for j in range(3):
            count += img_arr[x+i][y+j][z]
            
    count = int(count/9)
    return(count)
```

```python
for i in range(height-2):
    for j in range(width-2):
        img_arr_2[i+1][j+1][0] = okno(i, j, 0)
        img_arr_2[i+1][j+1][1] = okno(i, j, 1)
        img_arr_2[i+1][j+1][2] = okno(i, j, 2)    
            
Image.fromarray(img_arr_2, 'RGB')
```
