```python
#Изменение контраста +-
import math
import numpy as np
from PIL import Image

img = Image.open('4Labs.jpg')
img_arr = np.array(img)
img_arr_2 = img_arr

const=100

def pixel_contrast(x, level):
    factor = (259 * (level + 255)) / (255 * (259 - level))
    value = 128 + factor * (x - 128)
    return max(0, min(255, value))

def contrst(x, y, z):
    result = (int(pixel_contrast(x, const)), int(pixel_contrast(y, const)), int(pixel_contrast(z, const)))
    return result
    
for i in range(height):
    for j in range(width):
        img_arr_2[i][j] = contrst(img_arr[i][j][0], img_arr[i][j][1], img_arr[i][j][2])
                    
Image.fromarray(img_arr_2, 'RGB')
```

```python
#RGB to HSV 

import math
import numpy as np
from PIL import Image

img = Image.open('4Labs.png')
img_arr = np.array(img)
img_arr_2 = img_arr
width, height = img.size

for i in range(height):
    for j in range(width):
        r = img_arr[i][j][0]/255
        g = img_arr[i][j][1]/255
        b = img_arr[i][j][2]/255
        
        Cmax = max(r, g, b)
        Cmin = min(r, g, b)
        
        delta = Cmax-Cmin
        
        #Рассчет оттенка
        if delta == 0:
            img_arr_2[i][j][0] = 0
        elif Cmax == r:
            img_arr_2[i][j][0] = 60*(((g-b)/delta)%6)
        elif Cmax == g:
            img_arr_2[i][j][0] = 60*(((g-b)/delta)+2)
        else:
            img_arr_2[i][j][0] = 60*(((g-b)/delta)+4)
            
        #Рассчет насыщености    
        if Cmax == 0:
            img_arr_2[i][j][1] = 0
        else:
            img_arr_2[i][j][1] = delta/Cmax
            
        #Расчет стоимости    
        img_arr_2[i][j][2] = Cmax
            
            
```

```python
# HSV to RGB
import math
import numpy as np
from PIL import Image

img = Image.open('4Labs.png')
img_arr = np.array(img)
width, height = img.size

for i in range(height):
    for j in range(width):
        C = img_arr_2[i][j][1]*img_arr_2[i][j][2]
        X = C * (1 - abs((img_arr_2[i][j][0]/60)%2-1))
        m = img_arr_2[i][j][2] - C
        if 0 <= img_arr_2[i][j][0] < 60:
            R = C
            G = X
            B = 0
        elif 60 <= img_arr_2[i][j][0] < 120:
            R = X
            G = C
            B = 0
        elif 60 <= img_arr_2[i][j][0] < 180:
            R = 0
            G = C
            B = X
        elif 60 <= img_arr_2[i][j][0] < 240:
            R = 0
            G = X
            B = C
        elif 60 <= img_arr_2[i][j][0] < 300:
            R = X
            G = 0
            B = C
        else:
            R = C
            G = 0
            B = X
            
        img_arr_2[i][j][0] = (R+m)*255 
        img_arr_2[i][j][1] = (G+m)*255
        img_arr_2[i][j][2] = (B+m)*255
        
```
